1. 

Bemutatja, hogy a nettóárak miként helyezkednek al a megadott 4000, illetve 5000 értékhez viszonyulva.

```sql
DECLARE 
@min int = 4000,
@max int = 5000;   

SELECT 
    SUM(CASE WHEN nettoar < @min THEN 1 ELSE 0 end) AS '4000 FT alatt', 
    SUM(CASE WHEN nettoar BETWEEN @min and @max THEN 1 ELSE 0 end) AS '4000 és 5000 FT között',
    SUM(CASE WHEN nettoar > @max THEN 1 ELSE 0 end) AS '5000 Ft felett'
FROM dvd
```

2.

Megmutatja a nettóáraknak a maximumát, minimumát, átlagát, szórását, szórás nyégyzetét

```sql
SELECT MIN(nettoar) AS 'Minimum éték', 
       MAX(nettoar) AS 'Maximum éték', 
       round(AVG(nettoar), 0) AS 'Átlag kerekítve', 
       round(STDEV(nettoar), 0) AS 'Szórás kerekítve'
FROM dvd
WHERE nettoar IS NOT NULL
```

3.

Hány dvd-nél adtak meg nyelvet, illetve hányvál nem

```sql
Select count(Case when lang is not null then 1 end) as 'Meg van adva nyelv', Count(*)-count(case when lang is not null then 1 end) as 'Nincs megadott nyelv'
from dvd
```

4.

A leghosszab cím hossza

```sql
SELECT top 1 LEN(cim) AS 'Karakterek hossza',
       count(*) AS 'Előfordulása'
FROM dvd
GROUP BY LEN(cim)
ORDER BY len(cim) desc
```

5.

Stílusok darabszáma

```sql
select stilus, count(*) as 'DB'
from dvd
where stilus is not null
GROUP BY stilus
```

6.

TOP 10 legdrágább történekólmi film

```sql
SELECT top 10 cim, nettoar
FROM dvd
where  stilus = 'történelem'
GROUP by cim, nettoar
order by nettoar desc
```
